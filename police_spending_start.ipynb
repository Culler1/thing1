```python
import pandas as pd
```

```python
p_s18=pd.read_excel("new_police_spending/StatewideMuniAfr550.xls")
p_s17=pd.read_excel("new_police_spending/StatewideMuniAfr2017.xls")
p_s16=pd.read_excel("new_police_spending/StatewideMuniAfr2016.xls")
p_s15=pd.read_excel("new_police_spending/StatewideMuniAfr2015.xls")
p_s14=pd.read_excel("new_police_spending/StatewideMuniAfr2014.xls")
p_s13=pd.read_excel("new_police_spending/StatewideMuniAfr2013.xls")
p_s12=pd.read_excel("new_police_spending/StatewideMuniAfr2012.xls")
p_s11=pd.read_excel("new_police_spending/StatewideMuniAfr2011.xls")
p_s10=pd.read_excel("new_police_spending/StatewideMuniAfr2010.xls")
p_s09=pd.read_excel("new_police_spending/StatewideMuniAfr2009.xls")
p_s08=pd.read_excel("new_police_spending/StatewideMuniAfr2008.xls")
```

```python
main_spending1=pd.concat([p_s08,p_s09,p_s10,p_s11,p_s12,p_s13,p_s14,p_s15,p_s16,p_s17,p_s18],sort=False)
```

```python
main_spending1.columns=main_spending1.columns.str.replace(" ",'_').str.replace("/\n","_",regex=True).str.replace("-",'_',regex=False).str.replace("/",'_',regex=False).str.lower()
```

```python
po_spending1=main_spending1[['municipality_id','reporting_year','police_expenditures']]
fiveyr_po=po_spending1[po_spending1.reporting_year>2013].groupby('municipality_id').police_expenditures.describe().round(2).reset_index().rename(columns={'mean':'av_5yr','50%':'median_5yr','std':'std_5yr','max':'most_5yr'}).drop(columns={'count',"min",'25%','75%'})
tenyr_po=po_spending1.groupby('municipality_id').police_expenditures.describe().round(2).reset_index().rename(columns={'mean':'av_10yr','50%':'median_10yr','std':'std_10yr','max':'most_10yr'}).drop(columns={'count',"min",'25%','75%'})
```

```python
t_spending1=main_spending1[['municipality_id','reporting_year','total_expenditures']]
fiveyr_t_sp=t_spending1[t_spending1.reporting_year>2013].groupby('municipality_id').total_expenditures.describe().round(2).reset_index().rename(columns={'mean':'av_t_5yr','50%':'median_t_5yr','std':'std_t_5yr','max':'most_t_5yr'}).drop(columns={'count',"min",'25%','75%'})
tenyr_t_sp=t_spending1.groupby('municipality_id').total_expenditures.describe().round(2).reset_index().rename(columns={'mean':'av_t_10yr','50%':'median_t_10yr','std':'std_t_10yr','max':'most_t_10yr'}).drop(columns={'count',"min",'25%','75%'})
```
